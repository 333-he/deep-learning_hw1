```python

def read_file(file_path='/content/train.csv',
              label=1, is_train=True):
    """
    从CSV文件中读取特定标签的文本数据

    参数:
        file_path: CSV文件路径
        label: 要筛选的标签(1表示正面)
        is_train: 是否为训练数据(在此函数中可能不需要，保留以兼容)

    返回:
        包含指定标签文本的列表
    """
    # 读取CSV文件
    df = pd.read_csv(file_path)

    # 筛选指定标签的文本
    # 假设CSV中有'text'列存储文本，'target'列存储标签(1表示正面)
    texts = df[df['target'] == label]['text'].tolist()

    return texts
```

```python
import torch
from torch.utils.data import Dataset
import pandas as pd
from transformers import DistilBertTokenizer, DistilBertModel

class TweetDataset(Dataset):
    def __init__(self, csv_path, is_train=True, max_length=128):
        self.data = pd.read_csv(csv_path)
        self.is_train = is_train
        self.tokenizer = DistilBertTokenizer.from_pretrained('distilbert-base-uncased')
        self.max_length = max_length
    def __len__(self):
        return len(self.data)
    def __getitem__(self, idx):
        text = str(self.data.loc[idx, 'text'])
        encoded = self.tokenizer.encode_plus(
            text,
            add_special_tokens=True,
            max_length=self.max_length,
            padding='max_length',
            truncation=True,
            return_attention_mask=True,
            return_tensors='pt'
        )

        input_ids = encoded['input_ids'].squeeze(0)
        attention_mask = encoded['attention_mask'].squeeze(0)
        x = (input_ids, attention_mask)

        if not self.is_train:
          tweet_id = self.data.loc[idx, 'id'] if 'id' in self.data.columns else str(idx)
          return x, str(tweet_id)

        y = self.data.loc[idx, 'target']
        return x, torch.tensor(y, dtype=torch.long)
```

```python
from torch.nn.utils.rnn import pad_sequence

def get_dataloader(train_csv="/content/train.csv", test_csv="/content/test.csv", batch_size=32):
    def train_collate(batch):
    # batch = [ ( (input_ids, attention_mask), y ), ... ]
      x_list, y_list = zip(*batch)  # x_list = [ (input_ids, attention_mask), ... ]

    # 拆解 input_ids 和 attention_mask
      input_ids = [x[0] for x in x_list]  # List[Tensor(seq_len)]
      attention_masks = [x[1] for x in x_list]  # List[Tensor(seq_len)]

    # Pad input_ids 和 attention_mask
      input_ids_pad = pad_sequence(input_ids, batch_first=True)  # (batch_size, max_seq_len)
      attention_masks_pad = pad_sequence(attention_masks, batch_first=True)  # (batch_size, max_seq_len)

    # 合并成 (input_ids, attention_mask), y
      x_pad = (input_ids_pad, attention_masks_pad)
      y_pad = torch.tensor(y_list, dtype=torch.long)  # (batch_size,)

      return x_pad, y_pad
    def test_collate(batch):
    # batch = [ ( (input_ids, attention_mask), id ), ... ]
      x_list, ids = zip(*batch)  # x_list = [ (input_ids, attention_mask), ... ]

    # 拆解 input_ids 和 attention_mask
      input_ids = [x[0] for x in x_list]  # List[Tensor(seq_len)]
      attention_masks = [x[1] for x in x_list]  # List[Tensor(seq_len)]

    # Pad input_ids 和 attention_mask
      input_ids_pad = pad_sequence(input_ids, batch_first=True)  # (batch_size, max_seq_len)
      attention_masks_pad = pad_sequence(attention_masks, batch_first=True)  # (batch_size, max_seq_len)

    # 合并成 (input_ids, attention_mask), ids
      x_pad = (input_ids_pad, attention_masks_pad)

      return x_pad, list(ids)

    train_dataloader = DataLoader(
        TweetDataset(train_csv),
        batch_size=batch_size,
        shuffle=True,
        collate_fn=train_collate
    )
    test_dataloader = DataLoader(
        TweetDataset(test_csv, is_train=False),
        batch_size=batch_size,
        shuffle=False,
        collate_fn=test_collate
    )
    return train_dataloader, test_dataloader
```

```python
#import torch.nn as nn
class DisasterClassifier(nn.Module):
    def __init__(self, bert_model, num_classes=2, dropout_prob=0.1):
        super(DisasterClassifier, self).__init__()
        self.bert = bert_model
        self.dropout = nn.Dropout(dropout_prob)
        self.classifier = nn.Linear(self.bert.config.hidden_size, num_classes)
    def forward(self, input_ids, attention_mask):
        # 获取 BERT 的输出（确保是 BaseModelOutput）
        outputs = self.bert(
            input_ids=input_ids,
            attention_mask=attention_mask
        )
        # 取 [CLS] 标记的隐藏状态
        pooled_output = outputs.last_hidden_state[:, 0, :]  # (batch_size, hidden_size)
        pooled_output = self.dropout(pooled_output)
        return self.classifier(pooled_output)  # (batch_size, num_classes)



```
